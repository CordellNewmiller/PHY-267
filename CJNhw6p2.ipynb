```python
"""
         1         2         3         4         5         6         7      
1234567890123456789012345678901234567890123456789012345678901234567890123456

Cordell Newmiller
Prof. Lubin
PHY 267
Homework 6
Problem 2
"""

import numpy
from matplotlib import pyplot
%matplotlib inline

# Assumed import fields
# 3-5   RA    Right Ascension
# 6-8   Dec   Declination
# ???         Mass

data1 = numpy.genfromtxt('cluster1.dat',skip_footer=2)
data2 = numpy.genfromtxt('cluster2.dat',skip_footer=3,
                         usecols = (0,1,2,3,4,5,6,7,8))
data3 = numpy.genfromtxt('cluster3.dat',skip_footer=3,
                         usecols = (0,1,2,3,4,5,6,7,8))

def myunpacker(data):
    # Returns two column vector of x (RA) and y (Dec), both measured in degrees
    RAhour=data[:,3]
    RAmin=data[:,4]
    RAsec=data[:,5]
    DECdeg=data[:,6]
    DECmin=data[:,7]
    DECsec=data[:,8]
        
    position=numpy.zeros([data.shape[0],2])
    
    # Scale RA with declination
    position[:,0]=(RAhour*360/24 + RAmin * 360/(60*24) + RAsec*360/(60*60*24)
                  ) * (numpy.cos(numpy.mean(DECdeg) * numpy.pi/180))
    
    # Be aware of negative declinations
    position[:,1]=(numpy.abs(DECdeg) + DECmin /60 + DECsec /(60*60)
                  )*numpy.sign(DECdeg)
    
    return position


cluster1=myunpacker(data1)
cluster2=myunpacker(data2)
cluster3=myunpacker(data3)

def neighbordist(cluster,point):
    distances = numpy.sort((numpy.sum((cluster-point)**2,1))**(1/2))
    return numpy.mean(distances[0:5])


def myrotate(cluster,center,angle):
    # Expects angle in degrees
    
    xylist = cluster - center
    radials = (numpy.sum((xylist)**2,1))**(1/2)
    thetas = numpy.arctan2(xylist[:,1],xylist[:,0])
    
    thetaprimes = thetas + angle * numpy.pi/180
    xyprime=numpy.zeros(xylist.shape)
    xyprime[:,0]=radials*numpy.cos(thetaprimes)
    xyprime[:,1]=radials*numpy.sin(thetaprimes)
    
    return (xyprime + center)


def mydisplay(cluster,title):

    center = numpy.mean(cluster,0)
    radials = (numpy.sum((cluster-center)**2,1))**(1/2)
    
    fig=pyplot.figure(figsize=(10,5));
    fig.add_subplot(121)
    pyplot.plot(cluster[:,0],cluster[:,1],'+',center[0],center[1],'rx')
    pyplot.title(title)
    pyplot.xlabel('Right Ascension (deg)')
    pyplot.ylabel('Declination (deg)')    
    fig.add_subplot(122)
    pyplot.hist(radials,25)
    pyplot.title('Radial Density Profile')
    pyplot.xlabel('Radial Distance (deg)')
    pyplot.ylabel('Count')
    
    oppositepoints = 2*center - cluster
    
    # Average beta is a measure of asymmetry, with 0 being symmetrical
    # West et al. (1988), ApJ, 327, 1
    beta=numpy.zeros(radials.shape)
    for i in range(cluster.shape[0]):
        beta[i] = numpy.log10(neighbordist(cluster,oppositepoints[i,:])
                       /neighbordist(cluster,cluster[i,:]))
    
    print(title,'average beta is %2.3f ± %1.3f' %(beta.mean(), 
                        beta.std()/(beta.size)**(1/2)))
    
    spread=numpy.zeros([180,1])
    for i in range(180):
        rotclust=myrotate(cluster,center,i)
        spread[i]=numpy.std(rotclust[:,0])
    
    # I came up with this statistic to compute flatness, I wonder if 
    # it has a more accepted name already. 
    roundness = (numpy.max(spread)/numpy.min(spread))
    
    print('\tand its roundness is %2.2f \n' %(roundness))
    
    return

mydisplay(cluster1,'Cluster 1')
mydisplay(cluster2,'Cluster 2')
mydisplay(cluster3,'Cluster 3')

print('All three of these clusters have substructure, although 2 is the smoothest.')
print('All three of them are fairly flat, although 1 is the roundest.')

# End of problem 2
```
